```python
from pymc import gp
import numpy as np

# Generate mean
def quadfun(x, a, b, c):
    return (a*x**2 + b*x + c)

M = gp.Mean(quadfun, a=1., b=0.5, c=2.)

x = np.arange(-1,1,0.1)
plt.plot(x, M(x), 'k-')
```

```python
from pymc.gp.cov_funs import matern

C = gp.Covariance(eval_fun=matern.euclidean, diff_degree=3.5, amp=0.4, 
                  scale=1, rank_limit=1000)

plt.subplot(1,2,2)
plt.contourf(x, x, C(x,x).view(ndarray), origin='lower', extent=(-1,1,-1,1), 
            cmap=cm.bone)
plt.colorbar()

plt.subplot(1,2,1)
plt.plot(x, C(x,0).view(ndarray), 'k-')
plt.ylabel('C(x,0)')
```

```python
# Generate realizations
f_list = [gp.Realization(M, C) for i in range(3)]

# Plot mean and covariance
x = np.arange(-1,1,0.01)
gp.plot_envelope(M, C, x)

# Add realizations
for f in f_list:
    plt.plot(x, f(x))
```
